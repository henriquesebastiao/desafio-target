## Questão 4

Dado o valor de faturamento mensal de uma distribuidora, detalhado por estado:

- SP – R$67.836,43
- RJ – R$36.678,66
- MG – R$29.229,88
- ES – R$27.165,48
- Outros – R$19.849,53

Escreva um programa na linguagem que desejar onde calcule o percentual de representação que cada estado teve dentro do valor total mensal da distribuidora.

```python
import matplotlib.pyplot as plt

invoicing = {
    'sp': 67836.43,
    'rj': 36678.66,
    'mg': 29229.88,
    'es': 27165.48,
    'outros': 19849.53,
}

total = sum(invoicing.values())

labels = []
sizes = []

for state, value in invoicing.items():
    labels.append(state.upper())
    sizes.append((value / total) * 100)

colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#8b949e']
explode = (0, 0, 0, 0, 0)

plt.figure(figsize=(6, 6))
plt.pie(
    sizes,
    explode=explode,
    labels=labels,
    colors=colors,
    autopct='%1.2f%%',
    shadow=True,
    startangle=90,
)
plt.axis('equal')
plt.show()
```
